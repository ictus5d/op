```python
from sage.manifolds.operators import dalembertian
M = Manifold(2, 'M', structure='Lorentzian')
X.<t,x> = M.chart();g = M.metric()
g[0,0], g[1,1] = -1, 1
f = M.scalar_field((x-t)^3 + (x+t)^2, name='f')
Df = dalembertian(f);Df.display()
```
